```python
def onehot_encode(df, columns, prefixes):
  df = df.copy()
  for column, prefix in zip(columns, prefixes):
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df ,dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
df = onehot_encode(
    df,
    columns=['Beverage',' Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python
import numpy as np
import pandas as pd

# Replace 'Tall' with NaN
df = df.replace('Tall', np.nan)

# Convert values to numeric, handling '%' and non-numeric values
def convert_to_numeric(x):
  try:
    return np.float64(str(x).replace('%', ''))
  except ValueError:
    return np.nan

df = df.applymap(convert_to_numeric)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
numerical_df = df.select_dtypes(include=['number'])
```

```python
x = StandardScaler().fit_transform(numerical_df)
```

```python
# Applying PCA
pca = PCA()

pca.fit(x)
```

```python
# Check for columns with all NaN values
nan_columns = numerical_df.columns[numerical_df.isnull().all()]

# Remove those columns from your dataframe
numerical_df = numerical_df.drop(columns=nan_columns)

# Repeat the scaling and imputation steps
x = StandardScaler().fit_transform(numerical_df)

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)
```

```python
# Ploting variance ratio
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio')
plt.show()
```

```python
# Ploting variance ratio
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance Ratio')
plt.show()

# Choose the optimal number of components based on the plot above
# For example, select the number of components that explain 95% of the variance
optimal_num_components = 20 # Example value, adjust as needed

# Applying PCA with the optimal number of components
pca = PCA(n_components=optimal_num_components)

# Impute NaN values using the mean strategy
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
x = imputer.fit_transform(x)

principalComponents = pca.fit_transform(x)

# Create a DataFrame with the principal components
principalDf = pd.DataFrame(data = principalComponents, columns = [f'principal component {i+1}' for i in range(optimal_num_components)])

# Visualize the principal components
plt.figure(figsize=(8,6))
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Principal Components Visualization')
plt.show()

# Observations from the plot:
# - This plot shows the data projected onto the first two principal components.
# - You can observe how the data is distributed and if there are any distinct clusters or patterns.
# - Depending on the data and the explained variance ratio, you might need to visualize more principal components to get a better understanding of the data structure.
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

tsne = TSNE(n_components=2, random_state=42)
embeddings = tsne.fit_transform(x)

# Create a scatter plot of the embeddings
plt.figure(figsize=(8, 6))
plt.scatter(embeddings[:, 0], embeddings[:, 1])
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Visualization')
plt.show()
```
